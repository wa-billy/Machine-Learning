### VIF (Variance Inflation Factor) : ใช้ในการตรวจจับปัญหาความสัมพันธ์ร่วมหลายตัวแปรในการวิเคราะห์การถดถอย

```python
import pandas as pd

data = pd.read_csv('BMI.csv')

print(data)
```

```python
from statsmodels.stats.outliers_influence import variance_inflation_factor as vif

data['Gender'] = data['Gender'].map({'Male':0, 'Female':1})

X = data[['Gender', 'Height', 'Weight']]

vif_data = pd.DataFrame()
vif_data["feature"] = X.columns

vif_data["VIF"] = [vif(X.values, i) for i in range(len(X.columns))]

print(vif_data)
```

ค่า VIF สูงสำหรับส่วนสูงและน้ำหนักแสดงให้เห็นถึงความสัมพันธ์เชิงเส้นร่วมกันที่แข็งแกร่งระหว่างตัวแปรทั้งสอง ซึ่งเป็นเรื่องที่สมเหตุสมผลเนื่องจากส่วนสูงของบุคคลมีอิทธิพลต่อน้ำหนัก การตรวจจับความสัมพันธ์ดังกล่าวช่วยให้เราเข้าใจและปรับปรุงเสถียรภาพของแบบจำลองการถดถอยของเราได้ดียิ่งขึ้น
