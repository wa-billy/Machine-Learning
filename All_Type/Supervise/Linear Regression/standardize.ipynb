### Feature Scaling

- Normalization (Min - Max Scaling)

- คือ การปรับค่าช่วงข้อมูลให้มีค่าอยู่ระหว่าง 0 - 1 โดยรักษาความสัมพันธ์เชิงเส้นเดิมเอาไว้

```python
import numpy as np

X = np.array([100, 150, 200, 250, 300, 350, 400, 450, 500])

X_new = (X - min(X))/(max(X) - min(X))

print(X_new)
```

- Standardization (Z-score)

```python
X__new = (X - np.mean(X))/np.std(X)

print(X__new)
```
