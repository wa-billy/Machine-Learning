### Confusion Matrix 
- เป็นตารางอย่างง่ายที่ใช้ในการวัดประสิทธิภาพของแบบจำลองการจำแนกประเภท โดยจะเปรียบเทียบการคาดการณ์ของแบบจำลองกับผลลัพธ์จริง

### 1. Import Dependencies

```python
import numpy as np
from sklearn.metrics import confusion_matrix, classification_report
import seaborn as sns
import matplotlib.pyplot as plt
```

### 2. Create Numpy array for actual and predicted labels

```python
actual    = np.array(
  ['Dog','Dog','Dog','Not Dog','Dog','Not Dog','Dog','Dog','Not Dog','Not Dog'])
predicted = np.array(
  ['Dog','Not Dog','Dog','Not Dog','Dog','Dog','Dog','Dog','Not Dog','Not Dog'])
```

### 3. Compute the confusion metrix

```python
cm = confusion_matrix(actual, predicted)
```

### 4. Plot the confusion metrix with the help of the seaborn heatmap
- sns.heatmap:ฟังก์ชันนี้จากSeaborn ใช้สำหรับสร้างฮีทแมปของเมทริกซ์ความสับสน
- annot=True:แสดงค่าตัวเลขในแต่ละเซลล์ของฮีทแมป

```python

sns.heatmap(
    cm,
    annot=True,
    fmt='g',
    xticklabels=['Dog', 'Not Dog'],
    yticklabels=['Dog', 'Not Dog']
)
plt.ylabel('Actual', fontsize=13)
plt.title('Confusion Matrix', fontsize=17, pad=20)
plt.gca().xaxis.set_label_position('top')
plt.xlabel('Prediction', fontsize=13)
plt.gca().xaxis.tick_top()

plt.gca().figure.subplots_adjust(bottom=0.2)
plt.gca().figure.text(0.5, 0.05, 'Prediction', ha='center', fontsize=13)
plt.show()
```

### 5. Classification Report based on Confusion Metrix

```python
print(classification_report(actual, predicted))
```
