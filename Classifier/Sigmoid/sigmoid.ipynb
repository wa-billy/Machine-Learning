## Sigmoid Function

###
- ฟังก์ชันซิกมอยด์ (Sigmoid) เป็นฟังก์ชันทางคณิตศาสตร์ที่แปลงจำนวนจริงใดๆ ให้เป็นค่าระหว่าง 0 ถึง 1 ลักษณะเส้นโค้งรูปตัว "S" ทำให้ฟังก์ชันนี้มีประโยชน์อย่างยิ่งในสถานการณ์ที่เราต้องการแปลงผลลัพธ์เป็นความน่าจะเป็น ฟังก์ชันนี้มักถูกเรียกว่าฟังก์ชันโลจิสติก (Logistic function)

![alt text](<สกรีนช็อต 2025-12-24 174110.png>)

```python
import matplotlib.pyplot as plt
import numpy as np
import math

x = np.linspace(-10, 10, 100)
z = 1/(1 + np.exp(-x))

plt.plot(x, z)
plt.xlabel("x")
plt.ylabel("Sigmoid(x)")

plt.show()
```
