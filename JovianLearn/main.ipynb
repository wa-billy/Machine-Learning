```python
import pandas as pd

medical_df = pd.read_csv('medical.csv')

medical_df.info()
```

```python
medical_df.describe()
```

```python
import plotly.express as px # use for create interactive graph
import matplotlib
import matplotlib.pyplot as plt # use for create a graph "state-base interface"
import seaborn as sns

%matplotlib inline
```

```python
sns.set_style('darkgrid')
matplotlib.rcParams['font.size'] = 14
matplotlib.rcParams['figure.figsize'] = (10, 6)
matplotlib.rcParams['figure.facecolor'] = '#0000000'
```

```python
"""
อายุเป็นคอลัมน์ตัวเลข อายุต่ำสุดในชุดข้อมูลคือ 18 ปี และอายุสูงสุดคือ 64 ปี ดังนั้น 
เราสามารถแสดงภาพการกระจายของอายุโดยใช้ฮิสโตแกรมที่มี 47 ช่วง (หนึ่งช่วงสำหรับแต่ละปี) 
และแผนภาพกล่อง เราจะใช้ plotly เพื่อสร้างแผนภูมิแบบโต้ตอบ 
แต่คุณสามารถสร้างแผนภูมิที่คล้ายกันได้โดยใช้ Seaborn
"""

medical_df.age.describe()
```

```python
# Create a interactive graph to count age
fig = px.histogram(
    medical_df,
    x='age',
    marginal='box',
    nbins=47,
    title='Distribution of Age'
)
fig.update_layout(bargap=0.1)
fig.show()
```

```python
# BMI
fig = px.histogram(
    medical_df,
    x='bmi',
    marginal='box',
    color_discrete_sequence=['red'],
    title='Distribution of BMI (Body Mass Index)'
)
fig.update_layout(bargap=0.1)
fig.show()
```
